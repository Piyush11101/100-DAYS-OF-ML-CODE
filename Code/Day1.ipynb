# Importing Dataset

```python
dataset = pd.read_csv('/home/piyush/Desktop/Day1.csv')
```

```python
X = dataset.iloc[:, :-2].values
Y = dataset.iloc[:, -2].values
```

```python
print(X)
```

```python
print(Y)
```

# Handling the missing data

```python
#We should eliminate missing values by eliminating those rows but it is not effective when there are many missing rows
#Another method is to fill them with mean or median
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer = imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

```python
print(X)
```

#  ENCODING CATEGORICAL DATA

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

le = LabelEncoder()
X[:, 0] = le.fit_transform(X[:, 0])
```

```python
X
```

```python
# OneHotEncoding using columntransfer
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(categories='auto'), [0])],
    remainder='passthrough'
)

X = ct.fit_transform(X)
```

```python
print(X)
```

# SPLITTING THE DATASET 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=0)
```

# FEATURE SCALING

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)
```
